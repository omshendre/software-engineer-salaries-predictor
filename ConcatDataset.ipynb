```python
df_2021 = pd.read_csv('survey_results_public_2021.csv')
df_2022 = pd.read_csv('survey_results_public_2022.csv')
```

```python
def find_value_in_list(value, my_list):
    if value in my_list:
        return True
    else:
        return False

col_names_2021 = df_2021.columns.tolist()
col_names_2022 = df_2022.columns.tolist()

for col in col_names_2022:
    if not find_value_in_list(col, col_names_2021):
        print(col)
```
